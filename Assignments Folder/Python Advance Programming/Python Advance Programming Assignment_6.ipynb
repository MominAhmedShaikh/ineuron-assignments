### 1. You are given two strings s and t. String t is generated by randomly shuffling string s and then adding one more letter at a random position. Return the letter that was added to t.<br>
**Examples:**<br>
`find_the_difference("abcd", "abcde") ➞ "e"`<br>
`find_the_difference("", "y") ➞ "y"`<br>
`find_the_difference("ae", "aea") ➞ "a"`<br>

```python
def find_the_difference(in_one,in_two):
    in_two_temp = list(in_two)
    for ele in in_one:
        if ele in in_two_temp:
            in_two_temp.pop(in_two_temp.index(ele))
    output = ','.join(in_two_temp)
    print(f'find_the_difference{in_one,in_two} ➞ {output}')

find_the_difference("abcd", "abcde")
find_the_difference("", "y")
find_the_difference("ae", "aea")
```

### 2. Given a function that accepts unlimited arguments, check and count how many data types are in those arguments. Finally return the total in a list.<br>
`List order is: [int, str, bool, list, tuple, dictionary]`<br>
**Examples:**<br>
`count_datatypes(1, 45, "Hi", False) ➞ [2, 1, 1, 0, 0, 0]`<br>
`count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1) ➞ [3, 0, 0, 1, 1, 0]`<br>
`count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23) ➞ [2, 2, 3, 1, 0, 2]`<br>
`count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6]) ➞ [2, 0, 1, 2, 2, 0]`<br>

```python
from collections import OrderedDict

def count_datatypes(*args):
    # If nothing is entered returns [0,0,0,0,0,0]
    if not args:
        return [0, 0, 0, 0, 0, 0]

    # dict to store values as count 
    types = {int: 0, str: 0, bool: 0, list: 0, tuple: 0, dict: 0}
    
    od = OrderedDict([(int, 0), (str, 0), (bool, 0), (list, 0), (tuple, 0), (dict, 0)])

    for k, v in types.items():
        for a in args:
            if type(a) == k:
                od[k] += 1

    res = [i for i in od.values()]

    return f"count_datatypes({args}) ➞ {res}"
```

```python
print(count_datatypes(1, 45, "Hi", False))
print(count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1))
print(count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23))
print(count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6]))
```

### 3. A Fibonacci string is a precedence of the Fibonacci series. It works with any two characters of the English alphabet (as opposed to the numbers 0 and 1 in the Fibonacci series) as the initial items and concatenates them together as it progresses in a similar fashion as the Fibonacci series.<br>
**Examples:**<br>
`fib_str(3, ["j", "h"]) ➞ "j, h, hj"`<br>
`fib_str(5, ["e", "a"]) ➞ "e, a, ae, aea, aeaae"`<br>
`fib_str(6, ["n", "k"]) ➞ "n, k, kn, knk, knkkn, knkknknk"`<br>

```python
def fib_str(n, strings):
    if n <= 0:
        return ""
    elif n == 1:
        return strings[0]
    elif n == 2:
        return ", ".join(strings)
    
    fib_strings = [strings[0], strings[1]]
    for i in range(2, n):
        fib_strings.append(fib_strings[i-1] + fib_strings[i-2])
        
    return ", ".join(fib_strings)

print(fib_str(3,['j','h']))
print(fib_str(5, ["e", "a"]))
print(fib_str(6, ["n", "k"]))
```

### 4. Given an integer between 0 and 26, make a variable (self.answer). That variable would be assigned to a string in this format:<br>
`"nines:your answer, threes:your answer, ones:your answer"`<br><br>
`You need to find out how many ones, threes, and nines it would at least take for the number of each to add up to the given integer when multiplied by one, three or nine (depends).`<br>
**Examples:**<br>
`ones_threes_nines(10) ➞ "nines:1, threes:0, ones:1"`<br>
`ones_threes_nines(15) ➞ "nines:1, threes:2, ones:0"`<br>
`ones_threes_nines(22) ➞ "nines:2, threes:1, ones:1"`<br>

```python
def ones_threes_nines(n):
    m = n
    counter = {'nines':0,'threes':0,'ones':0}
    while n >= 9:  # n > 9
        n = n - 9
        counter['nines'] += 1
        if (n < 9) & (n >= 3): # 9 to =3
            n = n - 3
            counter['threes'] += 1
        if (n < 3) & (n >= 1): # 3 to =1
            counter['ones'] += 1
    return f"ones_threes_nines({n})  ➞ 'nines' : {counter['nines']}, threes' : {counter['threes']}, ones' : {counter['ones']}"
```

```python
print(ones_threes_nines(10))
print(ones_threes_nines(15))
print(ones_threes_nines(22))
```

### 5. The Fibonacci sequence is a classic use case for recursive functions since the value of the sequence at a given index is dependent on the sum of the last two values. However, the recursion tree created by solving the Fibonacci sequence recursively can grow quite fast. Therefore it can be important to think about the implications of running a function recursively. Depending on the size of n needed and the capabilities of the system in question you might want to take a different approach.<br>
`Write a non-recursive function that takes an integer n and returns the Fibonacci sequence's value at index n.`<br>
**Examples:**<br>
`fib(6) ➞ 8`<br>
`Output :  0 + 1 = 1, 1 + 1 = 2, 1 + 2 = 3, 2 + 3 = 5, 3 + 5 = 8` <br>
`fib(1) ➞ 1`<br>
`fib(2) ➞ 1`<br>

```python
# Appraoch 1

def fib1(x:int):
    if x == 1:
        return f"fib({x}) ➞ {1}"
    a = []
    for i,j in zip(range(1,x+2),range(0,x+2)):
        if j in [0,1]:
            a.append(j)
        if j > 1:
            a.append(a[j-1] + a[j-2])
    return f"fib({x}) ➞ {a[-1]}"
```

```python
print(fib1(6))
print(fib1(1))
print(fib1(2))
```

```python
# Appraoch 2

def fib2(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return f"fib({n}) ➞ {a}"
```
